```python
parser = Sbopen()
```

```python
df_wc_matches = parser.match(competition_id=43, season_id=106)
```

```python
df_wc_matches = df_wc_matches.sort_values('kick_off')
df_wc_matches
```

```python
parser.event(match_id=3857286)[0].head()
```

```python
players_xg_dct = {}
player_goals_dct = {}

for idx, match in df_wc_matches.iterrows():
    for idx2, event in parser.event(match.match_id)[0].iterrows():
        if event.type_name == 'Shot' and event.period < 5:
            players_xg_dct[event.player_name] = players_xg_dct.get(event.player_name,0) + event.shot_statsbomb_xg
            if event.outcome_name == 'Goal':
                player_goals_dct[event.player_name] = player_goals_dct.get(event.player_name,0) + 1
            else:
                player_goals_dct[event.player_name] = player_goals_dct.get(event.player_name,0) + 0
```

```python
len(players_xg_dct)
```

```python
len(player_goals_dct)
```

```python
players_xg_dct = dict(sorted(players_xg_dct.items(), key=lambda item: item[1], reverse=True))
player_goals_dct = dict(sorted(player_goals_dct.items(), key=lambda item: item[1], reverse=True))
```

```python
player_goals_dct
```

```python
players_xg_dct
```

```python
player_goals_df = pd.DataFrame(list(player_goals_dct.items()), columns=['Player', 'Goals'])
```

```python
#player_goals_df = player_goals_df[player_goals_df['Goals'] >= 0]
```

```python
player_goals_df['Expected_Goals'] = player_goals_df['Player'].apply(lambda player : round(players_xg_dct[player],2))
```

```python
player_goals_df['Diff'] = player_goals_df['Goals'] - player_goals_df['Expected_Goals'] 
```

```python
player_goals_df.sort_values('Diff', ascending = False)
```

```python
df_arg_matches = df_wc_matches[(df_wc_matches['home_team_name']=="Argentina") | (df_wc_matches['away_team_name']=="Argentina")]
```

```python
df_arg_matches
```

```python
alvarez_shots = pd.DataFrame()
for arg_match in df_arg_matches.match_id:
    temp_df = parser.event(arg_match)[0]
    temp_df = temp_df[(temp_df.type_name == 'Shot') & (temp_df.player_name == 'Julián Álvarez')]
    alvarez_shots = pd.concat([alvarez_shots,temp_df])
```

```python
pitch = Pitch(line_color='black')
fig, ax = pitch.draw(figsize = (10,8))

pitchLengthX = 120
pitchWidthY = 80

for i, shot in alvarez_shots.iterrows():
    x = shot['x']
    y = shot['y']
    
    circle_size = 2
    
    if shot['outcome_name'] == 'Goal':
        shot_circle = plt.Circle((x, y), circle_size, color='black')
    else:
        shot_circle = plt.Circle((x, y), circle_size, color='black')
        shot_circle.set_alpha(0.4)
    
    ax.add_patch(shot_circle)
fig.suptitle("Alvarez Shots in World Cup 2022", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```
