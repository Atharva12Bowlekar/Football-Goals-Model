```python
import pandas as pd
```

```python
url = "https://fbref.com/en/comps/9/2022-2023/schedule/2022-2023-Premier-League-Scores-and-Fixtures"
```

```python
epl_list = []


epl_list.append(pd.read_html(url,index_col=False,flavor='lxml')[0])
epl_list = pd.concat(epl_list, axis=0, ignore_index=True)


epl_df = epl_list[epl_list['Wk'].notna()]

epl_df = epl_df.rename(columns={'xG':'xGHome'
                   ,'xG.1':'xGAway'})


epl_df['HomeScore'] = epl_df['Score'].str[0]
epl_df['AwayScore'] = epl_df['Score'].str[2]

epl_df = epl_df.drop(['Match Report','Notes'],axis=1)

epl_df['Date'] = pd.to_datetime(epl_df['Date'])

epl_df.sort_values(by='Date', inplace=True)

epl_df = epl_df[['Wk','Day','Date','Time','Home','HomeScore','xGHome','AwayScore','xGAway',
                 'Away','Attendance','Venue','Referee']]
```

```python
epl_teams = list(epl_df['Home'].unique())
epl_teams.sort()
```

```python
epl_df.info()
```

```python
epl_df.tail(20)
```

```python
epl_df['HomeScore'] = epl_df['HomeScore'].apply(lambda x: int(x))
epl_df['AwayScore'] = epl_df['AwayScore'].apply(lambda x: int(x))
```

```python
epl_df.head(3)
```

```python
epl_teams_dict = {team: 0 for team in epl_teams}
xptd_goals_for = {team: 0 for team in epl_teams}
xptd_goals_agnst = {team: 0 for team in epl_teams}
```

```python
actual_points_dict = {team: 0 for team in epl_teams}
goals_for = {team: 0 for team in epl_teams}
goals_agnst = {team: 0 for team in epl_teams}
```

```python
def points_tally(row):
    if abs(row['xGHome'] - row['xGAway']) > 0.5:
        if row['xGHome'] > row['xGAway']:
            epl_teams_dict[row['Home']] += 3
        else:
            epl_teams_dict[row['Away']] += 3
    else:
        epl_teams_dict[row['Home']] += 1
        epl_teams_dict[row['Away']] += 1
        
    if row['HomeScore'] > row['AwayScore']:
        actual_points_dict[row['Home']] += 3
    elif row['HomeScore'] < row['AwayScore']:
        actual_points_dict[row['Away']] += 3
    else:
        actual_points_dict[row['Home']] += 1
        actual_points_dict[row['Away']] += 1
        
    xptd_goals_for[row['Home']] += row['xGHome']
    xptd_goals_for[row['Away']] += row['xGAway']
    xptd_goals_agnst[row['Home']] += row['xGAway']
    xptd_goals_agnst[row['Away']] += row['xGHome']
    
    goals_for[row['Home']] += row['HomeScore']
    goals_for[row['Away']] += row['AwayScore']
    goals_agnst[row['Home']] += row['AwayScore']
    goals_agnst[row['Away']] += row['HomeScore']
```

```python
epl_df['Result'] = epl_df.apply(points_tally, axis=1)
```

```python
epl_teams_dict = dict(sorted(epl_teams_dict.items(), key=lambda x: x[1], reverse=True))
```

```python
epl_xg_df = pd.DataFrame(data = list(epl_teams_dict.items()), columns=['Team', 'Expected_Points'])
```

```python
positions = list(range(1,21))
```

```python
epl_xg_df['Expected_Position'] = positions
```

```python
epl_xg_df
```

```python
def tallying_expected_goals(row):
    return round(xptd_goals_for[row['Team']],2), round(xptd_goals_agnst[row['Team']],2)
```

```python
epl_xg_df['Expected_Goals_For'] = epl_xg_df.apply(tallying_expected_goals, axis=1)
```

```python
epl_xg_df['Expected_Goals_Against'] = epl_xg_df['Expected_Goals_For'].apply(lambda x: x[1])
epl_xg_df['Expected_Goals_For'] = epl_xg_df['Expected_Goals_For'].apply(lambda x: x[0])
```

```python
actual_points_dict = dict(sorted(actual_points_dict.items(), key=lambda x: x[1], reverse=True))
actual_points_list = list(actual_points_dict.keys())
```

```python
def tallying_pos(row):
    return actual_points_list.index(row['Team']) + 1
```

```python
def tallying_goals_for(row):
    return goals_for[row['Team']]
```

```python
def tallying_goals_against(row):
    return goals_agnst[row['Team']]
```

```python
epl_xg_df['Actual_Position'] = epl_xg_df.apply(tallying_pos, axis=1)
epl_xg_df['Goals_For'] = epl_xg_df.apply(tallying_goals_for, axis=1)
epl_xg_df['Goals_Against'] = epl_xg_df.apply(tallying_goals_against, axis=1)
```

```python
def pos_change(row):
    return row['Actual_Position'] - row['Expected_Position']
```

```python
epl_xg_df['Position_Change'] = epl_xg_df.apply(pos_change, axis=1)
```

```python
epl_xg_df['Points'] = list(actual_points_dict.values())
```
